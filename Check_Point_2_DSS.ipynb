```python
import hvplot.pandas
import xarray as xr
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
performance_df = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
student_performance_df = pd.read_csv('StudentPerformanceFactors.csv')
```

```python
performance_df.head()
```

```python

plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Motivation_Level'].value_counts()
    for x_label, grp in performance_df.groupby('Extracurricular_Activities')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Atividades extracurriculares')
_ = plt.ylabel('Níveis de motivação')
```

```python
plt.figure(figsize=(14, 8))
sns.histplot(x='Exam_Score', hue='Learning_Disabilities', data=performance_df, multiple='dodge')
plt.title('Distribuição de notas por dificuldade de aprendizado')
plt.xlabel('Notas')
_ = plt.ylabel('Quantidade')
```

```python


plt.figure(figsize=(8, 6))
plt.scatter(performance_df['Motivation_Level'], performance_df['Hours_Studied'])
plt.title('Horas estudadas vs Nível de motivação')
plt.xlabel('Nível de motivação')
_ = plt.ylabel('Horas estudadas')
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(performance_df['Motivation_Level'], performance_df['Exam_Score'])
plt.title('Notas vs Nível de motivação')
plt.xlabel('Nível de motivação')
_ = plt.ylabel('Notas')
```

```python

performance_df['Hours_Studied'].plot(kind='hist', title='Horas de Sono')
plt.xlabel('Hours_Studied')
plt.ylabel('Motivation_Level')
plt.gca().spines[['top', 'right',]].set_visible(False)

```

```python
from matplotlib import pyplot as plt
performance_df['Sleep_Hours'].plot(kind='hist', bins=20, title='Horas de Sono')
plt.xlabel('Horas de Sono')
plt.ylabel('Frequência')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(performance_df['Sleep_Hours'], performance_df['Exam_Score'])
plt.title('Notas vs horas de sono')
plt.xlabel('Horas de sono')
_ = plt.ylabel('Notas')
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(performance_df['Hours_Studied'], performance_df['Exam_Score'])
plt.title('Notas vs horas de estudo')
plt.xlabel('Horas de estudo')
_ = plt.ylabel('Notas')
```
